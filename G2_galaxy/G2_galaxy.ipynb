```python
import h5py

def extract_parttype(f, parttype):  # the function definitions in this cell are from https://masterdesky.github.io/blog/astro/sim/gadget4-install
    '''
    Extracts the particle IDs, coordinates and velocities of a given
    particle type from a snapshot file.

    Parameters:
    -----------
    f : h5py.File
        The snapshot file opened with h5py.
    parttype : int
        The particle type to extract. 1 for dark matter (halo), 2 for
        disk particles.

    Returns:
    --------
    np.ndarray of shape (N, 7)
        A 2D numpy array with the particle IDs, coordinates and velocities.
    '''
    I = f[f'PartType{parttype}/ParticleIDs'][:]  # Particle IDs
    C = f[f'PartType{parttype}/Coordinates'][:]  # Particle Coordinates
    V = f[f'PartType{parttype}/Velocities'][:]   # Particle Velocities
    return np.hstack((I[:, np.newaxis], C, V))

def open_snapshot(s):
    '''
    Opens a GADGET-4 snapshot file and extracts the dark matter and disk
    particle data from it.

    Parameters:
    -----------
    s : str
        The path to the snapshot file.
    '''
    with h5py.File(s, 'r') as f:
        G_halo = extract_parttype(f, parttype=1)
        G_disk = extract_parttype(f, parttype=2)
        return G_halo, G_disk
```

```python
G_h0, G_d0=open_snapshot("data/snapshot_000.hdf5") # reading appropriate snapshots
G_h8, G_d8=open_snapshot("data/snapshot_080.hdf5")
G_h15, G_d15=open_snapshot("data/snapshot_015.hdf5")
```

```python
plt.plot(G_h0[:,1],G_h0[:,2], color = 'tab:blue',marker='.',markersize=1, linestyle='', label='Halo particles') # plotting and saving snapshots
plt.plot(G_d0[:,1],G_d0[:,2], color = 'tab:red',marker='.',markersize=1, linestyle='', label='Disk particles')
plot([],[], linestyle='', label="$t$ = 0 y")
leg=legend(loc='upper left')
xlabel('x [kpc/h]')
ylabel('y [kpc/h]', labelpad=-10)
leg.legend_handles[0].set_markersize(6)
leg.legend_handles[1].set_markersize(6)
grid(alpha=0.6, linestyle='--')
savefig('coll0.pdf', dpi=70, bbox_inches='tight')
```
